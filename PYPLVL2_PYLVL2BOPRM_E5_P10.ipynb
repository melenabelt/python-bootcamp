### **OPERADORES ARITMÉTICOS**
En python como en cualquier otro lenguaje es posible efectuar operaciones matemáticas como lo son la suma, resta, etc. A continuación veremos cada una de estas y como efectuarlas.

#### **SUMA**
En Python, + es el operador con el que se efectúa la operación de adición. Se utiliza para sumar 2 valores. En el siguiente ejemplo, puede verse como pueden sumarse dos valores.

**EJEMPLO**
```python
valor1 = 2
valor2 = 3

# Suma de valor1 + valor2
resultado = valor1 + valor2
5
```

#### **RESTA**
En Python, – es el operador de resta. Se utiliza para restar el segundo valor del primer valor.

**EJEMPLO**
```python
valor1 = 3
valor2 = 2

# Resta de valor1 - valor2
resultado = valor1 - valor2
1
```

#### **MULTIPLICACIÓN**
En Python, * es el operador de multiplicación. Se utiliza para encontrar el producto de 2 valores.

**EJEMPLO**
```python
valor1 = 3
valor2 = 2

# Multiplicación de valor1 * valor2
resultado = valor1 * valor2
6
```
#### **DIVISIÓN ENTERA**
En Python, el operador // se usa para realizar la división entre dos números y solo conservar la parte entera. Se usa para encontrar la parte entera del cociente cuando el primer operando se divide por el segundo.

**EJEMPLO**
```python
valor1 = 3
valor2 = 2

# División de valor1 // valor2
resultado = valor1 // valor2
1
```

#### **DIVISIÓN NORMAL**
En Python, / es el operador de división. Se utiliza para encontrar el cociente cuando el primer operando se divide por el segundo.

**EJEMPLO**
```python
valor1 = 3
valor2 = 2

# División de valor1 / valor2
resultado = valor1 / valor2
1.5
```

#### **EXPONENCIAL**
En Python, ** es el operador de exponenciación. Se utiliza para elevar el primer operando a la potencia del segundo.

**EJEMPLO**
```python
valor1 = 2
valor2 = 3

# Usando el operador de exponenciación
resultado = valor1 ** valor2
8
```

#### **MÓDULO**
En Python, % es el operador de módulo. Se utiliza para encontrar el resto o el sobrante cuando el primer operando se divide por el segundo.

**EJEMPLO**
```python
valor1 = 3
valor2 = 2

# Usando el operador de módulo
resultado = valor1 % valor2
1
```

#### **ORDEN DE OPERACIONES**
Pueden usarse los parentesis () al igual que en matemáticas, para dar prioridad a ciertas operaciones. Por ejemplo:
```python
operacion = (10 - 5) * 7 / 3 + 5
16.666666666666664
```
Es diferente a tener
```python
operacion = 10 - (5 * 7) / 3 + 5
3.333333333333334
```

Debido a que la prioridad de las operaciones ha sido modificada.


El orden de prioridad sería el siguiente para los operadores aritméticos, siendo el primero el de mayor prioridad:

 - () Paréntesis
 - ** Exponente
 - -x Negación
 - * / // Multiplicación, División, Cociente, Módulo
 - + - Suma, Resta

### Operadores Bitwise ó bit a bit

Los operadores bit a bit en Python se utilizan para realizar operaciones a nivel de bits. Permiten manipular bits individuales de enteros y realizar operaciones como AND, OR, XOR, desplazamientos y complementos. Estos operadores son particularmente útiles en programación de bajo nivel, redes y ciertos cálculos matemáticos.

Los operadores bitwise son: `&` (AND), `|` (OR), `^` (XOR), `~` (NOT), `<<` (Desplazamiento a la izquierda), `>>` (Desplazamiento a la derecha)

Es importante tener en cuenta que los operadores bit a bit trabajan en los bits individuales de los enteros, lo que los hace eficientes para ciertas operaciones bit a bit, pero su comprensión y uso requieren una buena comprensión de la representación binaria y cómo se manipulan los bits.

- Operador `&` (AND):<br/>
  Establece cada bit en 1 si ambos bits son 1.

- Operador `|` (OR):<br/>
  Establece cada bit en 1 si al menos uno de los bits es 1.

- Operador `^` (XOR):<br/>
  Establece cada bit en 1 si solo uno de los bits es 1.

- Operador `~` (NOT):<br/>
  Invierte todos los bits.

- Operador `<<` (Desplazamiento a la izquierda): <br/>
  Desplaza los bits hacia la izquierda, rellenando con ceros desde la derecha.

- Operador `>>` (Desplazamiento a la derecha): <br/>
  Desplaza los bits hacia la derecha, rellenando con ceros o el bit de signo desde la izquierda.
