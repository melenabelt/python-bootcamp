### **QUÉ ES FUNCIÓN**
Una función es un bloque de código que sólo se ejecuta cuando se llama. Pueden pasarse datos, conocidos como parámetros, a una función. Una función puede devolver datos como resultado.

### **CREANDO UNA FUNCIÓN**
En Python una función se define usando la palabra clave def:
  ```python
     def my_function():
        print("Hola soy una función")
  ```

### **LLAMANDO UNA FUNCIÓN**
Para llamar a una función, use el nombre de la función seguido de paréntesis:
  ```python
     def my_function():
        print("Hola soy una función")
  ```

### **ARGUMENTOS**
La información se puede pasar a funciones como argumentos. Los argumentos se especifican después del nombre de la función, dentro del paréntesis. Puedes agregar tantos argumentos como quieras, simplemente sepáralos con una coma. El siguiente ejemplo tiene una función con un argumento (fname). Cuando se llama a la función, pasamos un nombre, que se usa dentro de la función para imprimir el nombre completo:

  ```python
     def my_function(fname):
        print(fname + " Refsnes")

    my_function("Emil")
    my_function("Tobias")
    my_function("Linus")
  ```

### **¿ARGUMENTOS Ó PARÁMETROS?**
Los términos parámetro y argumento se pueden usar para lo mismo: información que se pasa a una función.

Desde la perspectiva de una función:

  * Un parámetro es la variable que figura entre paréntesis en la definición de la función.
  * Un argumento es el valor que se envía a la función cuando se llama.

````python
def function(inputs):

  FILTERED_OK = 0
  FILTERED_BAD = 0
  lista_keys = []
  lista_keys_uniques = []
  lista_keys_duplicadas = []
  padre_hijo: list = []

  diccionario = {
        "KEYS": [],
        "DATA": {
            "UNIQUE": [],
            "REPEATED": [],
            "ROOT": []
        }
    }


  for input in inputs:
    if isinstance(input, dict):
      FILTERED_OK += 1
      lista_keys.extend(input.keys())

      for key, value in input.items():
        if isinstance(value, dict):
          lista_keys.extend(value.keys())
          for llave, valor in value.items():
            padre_hijo.append(''.join(key + "." + llave))

    else:
      FILTERED_BAD += 1

  for elemento in lista_keys:
    if lista_keys.count(elemento) > 1:
        lista_keys_duplicadas.append(elemento)
    else:
        lista_keys_uniques.append(elemento)

  diccionario["KEYS"] = sorted(lista_keys)
  diccionario["DATA"]["UNIQUE"] = sorted(list(set(lista_keys_uniques)))
  diccionario["DATA"]["REPEATED"] = sorted(list(set(lista_keys_duplicadas)))
  diccionario["DATA"]["ROOT"] = sorted(padre_hijo)
  diccionario["DATA"]["LEN"] = len(inputs)
  diccionario["DATA"]["FILTERED_OK"] = FILTERED_OK
  diccionario["DATA"]["FILTERED_BAD"] = FILTERED_BAD

  print(diccionario)
  return (diccionario)


function([1.5, 2, 3, 'texto', {'nombre': 'Juan', 'edad': 30}])
````

El **FACTORIAL** de un número entero no negativo n, denotado como n!, es el producto de todos los números enteros positivos desde 1 hasta n. En otras palabras, el factorial de n se calcula multiplicando todos los números enteros positivos en orden, desde 1 hasta n.

La notación de factorial se ve así: `n!`, donde `n` es el número para el cual se está calculando el factorial. Aquí hay algunos ejemplos:

* El factorial de 0 es 1, es decir, `0! = 1`.
* El factorial de 1 es 1, es decir, `1! = 1`.
* El factorial de 2 es 2, es decir, `2! = 2 * 1 = 2`.
* El factorial de 3 es 6, es decir, `3! = 3 * 2 * 1 = 6`.
* El factorial de 4 es 24, es decir, `4! = 4 * 3 * 2 * 1 = 24`.

Los factoriales se utilizan en matemáticas y ciencias para calcular permutaciones, combinaciones y en diversas aplicaciones de conteo. También son utilizados en algoritmos y programación para resolver una variedad de problemas, como cálculos de probabilidades, análisis combinatorios entre otros.
